### **Entrada de valores para n,L e p**

```python
n_final = int(input('Insira o valor de n (n > 1): '))
L = int(input('Insira o valor de L (limite superior): '))
p = int(input('Insira o valor de p (passo tipo int): '))

t0 = -2
```

### Declaração variaveis globais e valroes constantes

```python
t = np.linspace(t0, L, p)
A0 = (e**20 - e**12 + e**8 -1) / (16 * e**8)
```

### Valor An e Bn calculado

```python
def An1(n):
  return (1/2)* ( ((e**12 * ((pi*n*sin(pi*n)) + 8*cos(pi*n)) -8*e**4 ) / (pi**2 * n**2 +64)) + ((2*e**-8 * (pi*n*sin(pi*n) - 8*cos(pi*n) + 8*e**8)) / (pi**2*n**2 + 64)) )
```

```python
def Bn1(n):
  return (1/2)* ( (-((2*e**4)*(8*e**8*sin(pi*n) - e**8 * pi*n *cos(pi*n) + pi*n)) / (pi**2 * n**2 +64)) - ((2*e**(-8)) * (8*sin(pi*n) + pi*n * (cos(pi*n) - e**8)) / (pi**2 * n**2 +64)) )
```

### Série exponencial de Fourier

```python
def somaSF(N, t):
  soma = np.zeros(np.size(t))
  for n in np.arange(1, N+1):
    soma +=  An1(n) * cos(pi/2*n*t) + Bn1(n) * sin(pi/2*n*t)
  return (A0/2 + soma)
```

### Plotagem do Gráfico

```python
def grafico(passo,y, titulo):
    plt.ioff()
    fig = plt.figure(figsize=(15, 8))
    plt.title(titulo)
    plt.grid(True)
    plt.ylabel('f(x)')
    plt.plot(passo,y)
    plt.show()
```

```python
y = []
aux = 0
titulo1 = "Gráfico 1"
for i in range(len(t)):
    x = np.array(somaSF(n_final, t[i]))
    y.append(x)
grafico(t,y, titulo1)
```

### Valor An e Bn calculados para n par e n impar

```python
def An31(n):
  return ((pi*n*sin(pi*n/2)) + (2*cos(pi*n/2)) - 2) / (2*pi*n**2)
```

```python
def An32(n):
  return  (2*cos(pi*n/2) + (-1)**n -3) / (2*pi*n**2) 
```

```python
def Bn31(n):
  return ((2*sin(pi*n/2) - pi*cos(pi*n/2) - sin(2*pi*n)) / (2*pi*n**2)) + ((-1)**n / (2*n))
```

```python
def Bn32(n):
  return (-sin(2*pi*n) + 2*sin(pi*n/2)) / (2*pi*n**2)
```

### Série de Fourier

```python
def somaSF(N, t):
  soma = np.zeros(np.size(t))
  for n in np.arange(1, N+1):
    if n % 2 == 0:
      soma +=  (A03/2) + An31(n)*cos(n*t) + Bn31(n)*sin(n*t)
    else:
      soma +=  (A03/2) + An32(n)*cos(n*t) + Bn32(n)*sin(n*t)
  return (A03/2) +soma
```
