### **Entrada de valores para n,L e p**

```python
n_final = int(input('Insira o valor de n (n > 1): '))
L = int(input('Insira o valor de L (limite superior): '))
p = int(input('Insira o valor de p (passo tipo int): '))

t0 = -2
```

### Declaração variaveis globais e valroes constantes

```python
t = np.linspace(t0, L, p)
# A0 = 186.3098532
```

### Valor Dn calculado

```python
def Dn(n):
  # (e**(pi*n+8) - (e**(-pi*n-8))) / (2*(pi*n + 8))
  return (((cos(pi*n+8) + sin(pi*n+8) - (cos(-pi*n-8) + sin(-pi*n-8))) / (2*(pi*n + 8))))
```

### Série exponencial de Fourier

```python
def somaSF(N, t):
  soma = np.zeros(np.size(t))
  for n in np.arange(1, N+1):
    soma += Dn(n) * (cos(n*t) + sin(n*t))
  return soma
```

### Plotagem do Gráfico

```python
def grafico(passo,y, titulo):
    plt.ioff()
    fig = plt.figure(figsize=(15, 8))
    plt.title(titulo)
    plt.grid(True)
    plt.ylabel('f(x)')
    plt.plot(passo,y)
    plt.show()
```

```python
y = []
aux = 0
titulo1 = "Gráfico 1"
for i in range(len(t)):
    x = np.array(somaSF(n_final, t[i]))
    y.append(x)
grafico(t,y, titulo1)
```

### Valor An e Bn calculados para n par e n impar

```python
def An2(n):
  #(pi*(-e**(-pi*n)+1)) / (2*n)
  return (2/ n**2)
```

```python
def Bn21(n):
  #((-pi**2 * n**2) -(2*pi*n) + (2*e**(pi*n) - 2)) / (-2*pi*n**3)
  return (((pi**2 * n**2) + (2*(-1)**n -2)) / (pi*n**3))
```

```python
def Bn22(n):
  #((-pi**2 * n**2) -(2*pi*n) + (2*e**(pi*n) - 2)) / (-2*pi*n**3)
  return (((pi**2 *(-1)**n * n**2) + (2*(-1)**n -2)) / (pi*n**3))
```

### Série de Fourier

```python
def somaSF(N, t):
  soma = np.zeros(np.size(t))
  for n in np.arange(1, N+1):
    if n % 2 == 0:
      soma +=  (A02/2) + An2(n)*cos(n*t) + Bn21(n)*sin(n*t)
    else:
      soma +=  (A02/2) + An2(n)*cos(n*t) + Bn22(n)*sin(n*t)
  return soma
```

```python
def An32(n):
  return  (2*cos(pi*n/2) + (-1)**n -3) / (2*pi*n**2) 
```

```python
def Bn31(n):
  return ((2*sin(pi*n/2) - pi*cos(pi*n/2) - sin(2*pi*n)) / (2*pi*n**2)) + ((-1)**n / (2*n))
```
